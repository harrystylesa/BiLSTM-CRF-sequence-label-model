```python
import pandas as pd
```

```python
df = pd.read_csv('../data/ner_pretrained_embedding/val_ner.txt_ids.csv', encoding = 'utf-8')
```

```python
def getint(l):
    ids = l[1:-1].replace(' ','').split(',')
#     print(ids)
    for i in range(len(ids)):
        if ids[i] in ['nan', 'NaN']:
            ids[i] = 0
        else:
            ids[i] = round(float(ids[i]))
    return ids
```

```python
df
```

```python
df['y_int'] = df['y'].apply(getint)
```

```python
labels = df['y_int'].to_list()
```

```python
df['X_int'] = df['X'].apply(getint)
```

```python
datas = df['X_int'].to_list()
```

```python
import tensorflow as tf
```

```python
writer = tf.python_io.TFRecordWriter('../data/ner_pretrained_embedding/val.tfrecords')
```

```python
def _int64_feature(value):
        """Returns an int64_list from a bool / enum / int / uint."""
        return tf.train.Feature(int64_list=tf.train.Int64List(value=value))
```

```python
for i in range(1710):
#                 label_list = list(map(lambda x: int(float(x)), labels[i]))
                example = tf.train.Example(features=tf.train.Features(feature=
                                                                      {'label': _int64_feature(labels[i]),
                                                                       'data': _int64_feature(datas[i])}))
                writer.write(example.SerializeToString())
```
